### Importance by order

```python
import plotly.express as px

def Importance_Order(c, n):
  # Get the correlations of all coins with c
  correlations = returns.corr()[c]
  
  # Get the n coins with the lowest and highest correlation to c
  lowest_corr_coins = correlations.sort_values(ascending=True)[:n]
  highest_corr_coins = correlations.sort_values(ascending=False)[1:1+n]
  
  # Combine the coins and their correlations into a dataframe
  coins_df = pd.DataFrame({
      'Coin': lowest_corr_coins.index.tolist() + highest_corr_coins.index.tolist(),
      'Correlation': lowest_corr_coins.values.tolist() + highest_corr_coins.values.tolist(),
      'Type': ['Lowest']*n + ['Highest']*n
  })
  
  # Sort the coins in descending order of correlation
  coins_df = coins_df.sort_values(by='Correlation', ascending=False)
  
  # Create the interactive bar chart using Plotly Express
  fig = px.bar(coins_df, x='Correlation', y='Coin', orientation='h', color='Type',
               color_discrete_sequence=['red', 'blue'], opacity=0.9)
  fig.update_traces(marker_line_width=0)
  fig.update_layout(title=f'Correlations of {n} coins with {c}', xaxis_title='Correlation')
  fig.show()
```

```python

def Rolling_corr(c1: str, c2: str, n : int =30) -> None:
    # Calculate the rolling correlation between the two cryptocurrencies
    rolling_corr = returns[c1].rolling(n).corr(returns[c2])[n:]
    
    # Create an interactive line chart of the rolling correlation using Plotly Express
    fig1 = px.line(x=rolling_corr.index, y=rolling_corr.values, title=f'Rolling Correlation of {c1} and {c2} over {n} days')
    fig1.update_xaxes(title='Time', showgrid=True, gridwidth=1, gridcolor='LightGray', tickfont=dict(size=12))
    fig1.update_yaxes(title='Rolling Correlation', showgrid=True, gridwidth=1, gridcolor='LightGray', tickfont=dict(size=12))
    fig1.update_traces(line=dict(color='steelblue', width=2))
    fig1.update_layout(title=dict(font=dict(size=20)), plot_bgcolor='white')
    fig1.show()
    
    # Create an interactive line chart comparing the returns of the two cryptocurrencies using Plotly Express
    fig2 = px.line(x=returns.index, y=[returns[c1], returns[c2]], title=f'Comparison of Returns for {c1} and {c2}')
    fig2.update_xaxes(title='Time', showgrid=True, gridwidth=1, gridcolor='LightGray', tickfont=dict(size=12))
    fig2.update_yaxes(title='Returns', showgrid=True, gridwidth=1, gridcolor='LightGray', tickfont=dict(size=12))
    fig2.update_traces(line=dict(width=2))
    fig2.update_layout(title=dict(font=dict(size=20)), plot_bgcolor='white')
    fig2.show()
```

```python

def Report(c, c2,returns=returns):
    
    # total number of days
    total_days = len(returns)
    
    # number of days on which both currencies gave positive returns
    return_earn = len(returns[(returns[c] > 1) & (returns[c2] > 1)])
    
    # number of days on which both currencies gave negative returns
    return_loss = len(returns[(returns[c] < 1) & (returns[c2] < 1)])
    
    # number of days on which both currencies gave returns of the same sign (positive or negative)
    same_direction = len(returns[((returns[c] > 1) & (returns[c2] > 1)) | ((returns[c] < 1) & (returns[c2] < 1))])
    
    # number of days on which both currencies gave returns of opposite sign (one positive, one negative)
    opposite_direction = len(returns) - same_direction
    
    # total returns for each currency
    total_returns_c = np.product(returns[c])
    total_returns_c2 = np.product(returns[c2])
    
    # average daily return for each currency
    avg_return_c = np.mean(returns[c])
    avg_return_c2 = np.mean(returns[c2])
    
    # standard deviation of daily returns for each currency
    std_return_c = np.std(returns[c])
    std_return_c2 = np.std(returns[c2])
    
    # correlation between daily returns of the two currencies
    correlation = np.corrcoef(returns[c], returns[c2])[0,1] 

    same_direction = return_earn + return_loss
    same_direction_pct = same_direction / len(returns) * 100

    # print report
    print(f"Total number of days: {total_days}")
    print(f"Number of days on which both currencies gave positive returns: {return_earn}")
    print(f"Number of days on which both currencies gave negative returns: {return_loss}")
    print(f"Number of days on which both currencies gave returns of the same sign: {same_direction}")
    print(f"Number of days on which both currencies gave returns of opposite sign: {opposite_direction}")
    print(f"Average daily return for {c}: {avg_return_c:.4f}")
    print(f"Average daily return for {c2}: {avg_return_c2:.4f}")
    print(f"Standard deviation of daily returns for {c}: {std_return_c:.4f}")
    print(f"Standard deviation of daily returns for {c2}: {std_return_c2:.4f} \n")
    print(f"Gold and silver went in the same direction on {same_direction_pct:.2f}% of days")
    print(f"Correlation between daily returns of {c} and {c2}: {correlation:.4f} ")
    
    
```

## BETA

In finance, beta is a measure of a stock's or other asset's volatility in relation to the overall market. It is often used as an indicator of the systematic risk or market risk associated with an investment.

---

The result of this code is the beta coefficient of the linear regression model, 

---

which is a measure of the sensitivity of the cryptocurrency's returns to the returns of the S&P 500 market index.

---


1.   A beta coefficient of 1 means that the cryptocurrency's returns move in line with the market index returns
2.   while a beta coefficient greater than 1 indicates that the cryptocurrency's returns are more volatile than the market index returns
3.   beta coefficient less than 1 indicates that the cryptocurrency's returns are less volatile than the market index returns.



```python
from sklearn.linear_model import LinearRegression

def Beta(crypto):
  # Calculate daily returns for each cryptocurrency and for the market (e.g. using a market index)
  crypto_returns = returns

  # Remove missing values
  market_returns.dropna(inplace=True)
  crypto_returns.dropna(inplace=True)

  # Calculate beta using linear regression
  model = LinearRegression()
  X = np.array(merged_data['market_Ret']).reshape(-1, 1)
  y = np.array(merged_data[crypto])
  model.fit(X, y)
  beta = model.coef_[0]
  return beta

Beta('BTC-USD')
```

```python
volatility = {}
for t in tickers:
    beta = Beta(t)
    volatility[t] = beta

sorted_volatility = sorted(volatility.items(), key=lambda x: x[1] if x[1] is not None else 0)

sorted_volatility = pd.DataFrame( sorted_volatility)
sorted_volatility.columns = ['Ticker', 'Volatility']

fig = px.bar(sorted_volatility, x="Ticker", y="Volatility", color="Volatility", color_continuous_scale=["lightblue", "darkblue"],
             title="BETA by Ticker",
             labels={"Volatility": "BETA Score"},
             width=1600, height=800)

fig.show()
```

# SHARP RATIO

The Sharpe ratio is a way to measure the performance of an investment while taking into account the level of risk involved.

---

A higher Sharpe ratio means that the investment has provided a higher return per unit of risk, which means that it is generally considered to be a better investment compared to others with lower Sharpe ratios.

On the other hand, a lower Sharpe ratio suggests that the investment has provided a lower return for the amount of risk taken, so it is generally considered to be a riskier investment.

```python

# Calculate annualized average daily returns and annualized daily volatility
avg_daily_returns = returns.mean()
std_daily_returns = returns.std()
annualized_avg_returns = avg_daily_returns * 365
annualized_volatility = std_daily_returns * np.sqrt(365)

# Calculate risk-free rate (e.g. using a 10-year Treasury bond rate)
risk_free_rate = 0.0237 # it varies with the time

# Calculate Sharpe ratio
sharpe_ratio = (annualized_avg_returns - risk_free_rate) / annualized_volatility

sharpe_ratios = pd.DataFrame({"Ticker": tickers, "Sharpe Ratio": sharpe_ratio})

# Sort by Sharpe ratio in descending order
sorted_sharpe_ratios = sharpe_ratios.sort_values(by="Sharpe Ratio", ascending=False)

# Create a bar chart of Sharpe ratios by ticker
fig = px.bar(sorted_sharpe_ratios, x="Ticker", y="Sharpe Ratio", color="Sharpe Ratio", color_continuous_scale=["lightblue", "darkblue"],
             title="Sharpe Ratios by Ticker",
             labels={"Sharpe Ratio": "Sharpe Ratio Score"},
             width=1600, height=800)
fig.show()
```
